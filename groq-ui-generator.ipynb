### IMPORTING LIBRARIES

````python
#import necessary libraries
import os
from groq import Groq
from IPython.display import display, HTML
from dotenv import load_dotenv
````

### SETTING API KEY

````python
# Load environment variables from .env file
load_dotenv()

# Initialize the Groq client with the API key from environment variables
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
client = Groq(api_key=GROQ_API_KEY)
````

### GIVING The PROMPT FOR THE OUTPUT

````python
# Create a beautiful, modern, neon-themed digital clock using the Groq API
print("Groq is coding the app... blink and you'll miss it!")

#The model is prompted to generate raw HTML, CSS, and JavaScript code for the clock
ompletion = client.chat.completions.create(
    model="llama-3.3-70b-versatile", # the newest supported model
    messages=[
        {
            "role": "system",
            "content": "You are an expert frontend developer. You only output raw HTML, CSS, and JS code. Do not use markdown formatting blocks. Do not explain the code."
        },
        {
            "role": "user",
            "content": "Create a beautiful, modern, neon-themed digital clock. Use dark mode, glowing green text, and center it on the screen. Make it functional with JavaScript."
        }
    ]
)

# Extracting the code from the model's response
html_code = completion.choices[0].message.content
html_code = html_code.replace("```html", "").replace("```", "") # Cleans any accidental markdown

# Display the generated code in the notebook
display(HTML(html_code))
````
